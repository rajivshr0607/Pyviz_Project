```python
# imports
import panel as pn
pn.extension('plotly')
import plotly.express as px
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
import os
from pathlib import Path
from dotenv import load_dotenv
from panel.interact import interact
from panel import widgets
```

# Import Data

```python
# Import the CSVs to Pandas DataFrames
file_path = Path("Data/sfo_neighborhoods_census_data.csv")
sfo_data = pd.read_csv(file_path, index_col="year")

file_path = Path("Data/neighborhoods_coordinates.csv")
df_neighborhood_locations = pd.read_csv(file_path)
```

## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    #sfo_data = pd.read_csv(file_path, index_col="year")
    #sfo_data.head()
    number_housing_units = sfo_data.groupby('year')['housing_units'].mean()
    number_housing_units_min = number_housing_units.min()
    number_housing_units_max = number_housing_units.max()
    number_housing_units_std = number_housing_units.std()
    fig = plt.figure(figsize=(10,10))
    plot_housing = number_housing_units.plot.bar(title="Housing Units in Sanfransisco form 2010 to 2016.", 
                                             ylim=(number_housing_units_min-number_housing_units_std,
                                                   number_housing_units_max+number_housing_units_std), figsize=(10,5))
    plot_housing.set_xlabel("Year")
    plot_housing.set_ylabel("Housing Units.")
    plt.close(fig)
#     plt.show()

    return pn.pane.Matplotlib(fig, tight=True);
    #return pn.pane.Plotly(fig, tight=True);



def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    fig = plt.figure()
    avg_grsR_saleSQ = sfo_data[['sale_price_sqr_foot','gross_rent']].groupby(sfo_data.index).mean()
    plot_avg_gross_rent = avg_grsR_saleSQ['gross_rent'].plot.line(title= "Average Gross Rent in Sanfransisco", figsize=(10,5))
    plot_avg_gross_rent.set_xlabel("Year")
    plot_avg_gross_rent.set_ylabel("Gross Rent")
    plt.close(fig)

    return pn.pane.Matplotlib(fig, tight=True);

def average_sales_price():
    """Average Sales Price Per Year."""
    fig = plt.figure()
    
    avg_grsR_saleSQ = sfo_data[['sale_price_sqr_foot','gross_rent']].groupby(sfo_data.index).mean()
    plot_sale_per_unit =  avg_grsR_saleSQ["sale_price_sqr_foot"].plot(kind="line",
                                                                      title= "Average Sales Price Per Year in Sanfransisco",
                                                                      figsize=(10,5))
    plot_sale_per_unit.set_xlabel("Year")
    plot_sale_per_unit.set_ylabel("Sales Price per Year")
    plt.close(fig)

    return pn.pane.Matplotlib(fig, tight=True);


def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""

    neighborhood_df = sfo_data.groupby(['year','neighborhood']).mean()
    neighborhood_df.reset_index(inplace=True)
    neighborhood_plot = neighborhood_df.hvplot.line(x="year", y="sale_price_sqr_foot",
                                                    groupby="neighborhood", xlabel="YEAR",
                                                    ylabel="Avg. Price Per sq ft")                                                 
    return neighborhood_plot;

def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    fig = plt.figure()

    top_expensive = sfo_data.groupby('neighborhood').mean()
    top_expensive = top_expensive.sort_values("sale_price_sqr_foot",   
    axis=0,
    ascending=False,
    inplace=False,
    kind='quicksort',
    na_position='last',
    ignore_index=False,)
    top_expensive = top_expensive.head(10)
    top_expensive.reset_index(inplace=True)
    top_expensive_plot = top_expensive.hvplot.bar(x="neighborhood", y="sale_price_sqr_foot",
                                                  title="Top 10 Expensive Housing",rot=90).opts(yformatter="%.0f")
    
    return top_expensive_plot;

def parallel_coordinates():
    """Parallel Coordinates Plot."""
    top_expensive = sfo_data.groupby('neighborhood').mean()
    top_expensive = top_expensive.sort_values("sale_price_sqr_foot",   
    axis=0,
    ascending=False,
    inplace=False,
    kind='quicksort',
    na_position='last',
    ignore_index=False,)
    top_expensive = top_expensive.head(10)
    top_expensive.reset_index(inplace=True)
    
    para_coord = px.parallel_coordinates(data_frame=top_expensive,
    dimensions=("neighborhood","sale_price_sqr_foot","housing_units","gross_rent"),
    color="sale_price_sqr_foot",
    labels={},
    color_continuous_scale=None,
    range_color=None,
    color_continuous_midpoint=None,
    title="Top 10 Expensive Neighborhood.",
    template=None,
    width=1000,                                     
    height=None
    )
    return para_coord
    # YOUR CODE HERE!


def parallel_categories():
    """Parallel Categories Plot."""
    top_expensive = sfo_data.groupby('neighborhood').mean()
    top_expensive = top_expensive.sort_values("sale_price_sqr_foot",   
    axis=0,
    ascending=False,
    inplace=False,
    kind='quicksort',
    na_position='last',
    ignore_index=False,)
    top_expensive = top_expensive.head(10)
    top_expensive.reset_index(inplace=True)
    para_cate = px.parallel_categories (top_expensive, color="sale_price_sqr_foot")
    
    return para_cate
        

    
    
    
    # YOUR CODE HERE!


def neighborhood_map():
    """Neighborhood Map"""
    ave_neighborhood = sfo_data.groupby('neighborhood').mean()
    ave_neighborhood.reset_index(inplace=True)
    ave_neighborhood.rename(columns={"neighborhood": "Neighborhood"}, inplace=True)
    avg_joined = pd.merge(df_neighborhood_locations,ave_neighborhood, on="Neighborhood")
    
    map = px.scatter_mapbox(
    avg_joined,
    lat="Lat",
    lon="Lon",
    color="gross_rent",
    title="Scatter Visualization",
    hover_name="Neighborhood",
    size="sale_price_sqr_foot",
    
    )
    
    return map
    # YOUR CODE HERE!

```

```python
all_tabs= pn.Tabs(
    ("#Scatter_Map", pn.Column(
        "##Scatter Vizualization of San Fransisco neighborhood.",
        neighborhood_map
    )
    ),
    ("#Parallels", pn.Column(
        "##Parallel coordiante and categories for top10 expensive areas in Sanfransisco.",
        parallel_coordinates,
        parallel_categories
    )
    ),
    ("Averages", pn.Row(
        "## No.of housing units, average gross rent and average sales price in Sanfransisco.",
        housing_units_per_year,
        average_gross_rent,
        average_sales_price,
    )
    ),
    ("Neighborhoods", pn.Column(
        "## neighborhood",
        average_price_by_neighborhood,
        top_most_expensive_neighborhoods
    )
    )
)
panel_dashboard = pn.Row(" Dashboard Visualization of Sanfransisco.", all_tabs, width=1000)
```
